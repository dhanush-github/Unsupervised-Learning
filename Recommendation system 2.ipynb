> #### Q. What is a Recommender Engine?

A recommender engine functions as an application suggesting specific items such as movies, products, songs, etc., to users based on their interests, preferences, and browsing history.

Additionally, it may recommend items to a user by analyzing the similarity of their choices or characteristics (e.g., gender, age group, location) with those of other users on the platform.

### Problem Formulation -
---

> #### Q. How do we formulate a recommender system problem?

This problem is quite different from the ones that we've seen already i.e., Regression and Classification. Here we have to generate recommendations instead of predictions numbers or categories.

Now, let us assume that we have  $n$  users. $U_1$, $U_2$, $U_3$... upto $U_n$. Also, we have $m$ items. $I_1$, $I_2$, $I_3$... upto $I_m$.

**Users** -> $U_i$, where $i = 1, 2, 3... n$ </br>
**Items** -> $I_j$, where $j = 1, 2, 3... m$

- We have user $U_i$ who has read three Bookw.

### Data Representation -
---

> #### Q. How can we represent the data for a recommender engine?

The dataset is represented in the form of a **matrix**, lets call it `A`.

In `A`, each row represents a user $U_i$, and each column represents an item $I_j$. </br> This way, element $A_{ij}$ denotes user $U_i$'s interaction with product $I_j$.

</br>

This interaction can be measured in many ways, depending upon the problem in hand.
* If a users likes a video on TikTok.
* If a user subscribe to a YouTube channel.
* If a users marks a song as favorite on Spotify.

</br>

Basically, we try to represent whether the user showed any form of interest in that product.

Naturally, `A` is a `n x m` matrix, where `n` being the no. of rows (users) and `m` being the no. of columns (items).

## Cosine Similarity
---

#### **Intuition**

- The idea behind cosine similarity is if two words are similar, then their vector form will have small or zero angle b/w them.

- Let's consider example given below:
  - **eat** and **eating** are quite similar words, so their geometrical representation shows here angle is zero b/w them.
  - **eat** and **walk** are two different words, so they large angle b/w their vector respectivly.

#### **Dot Product**

- Let's say two words are represented in the form of x and y vector.
- Then, how do you take their dot product?
  - We multiply corresponding dim of a vector with the other vector and add them up

> #### Q. How can we store these similarity scores?

* We build a **similarity matrix** that represents how similar two items $I_i$ & $I_j$ or two users $U_i$ & $U_j$ are.

* Larger the values in the matrix, more similar are the corresponding items/users.

<img src='https://drive.google.com/uc?id=19FnSPkkCFmM6K19_Yactl1b526-k5eKf' width="900" height="500">


> #### Q. What are the drawbacks of this approach?

* One major problem with user-user similarity is that **User preferences can change over time**. This can lead to bad recommendations.

* In order to handle this, we prefer using item-item similarity approach, because in contrast, the ratings on given items do not change significantly over time.

> #### Q. How to decide when to use user-user or item-item similarity approach?

When we have more number of users than items, i.e. n > m, and if the item ratings do not change much over time, then it is better to use the item-item similarity approach.

### Cold Start Problem -
---

> #### Q. What if we have a new user joining the platform?

* When a new user joins, then we have no ratings given from this user, as he is new to the ecosystem. So, we cannot find similar users.

* For every recommender system, it is required to build a user profile by considering the user's activities and behaviours with the system.

* Based on user's history only, recommendations are made. If there is no historical data, that is a problem.



> #### Q. What if there is a new product that's added to the platform?

* Here again, we have no historical data, as it is a new product. We cannot find any similar items to it.






<img src='https://drive.google.com/uc?id=1DGRBFRp3BYUgCJGbj5X5J9t8SBJI1hUc' height="600">



Both these cases are called as a **Cold Start Problem**.

This problem arises due to 2 different reasons:-
- For new users
- For new items

In these cases, we don't have enough information to make good recommendations.

```python
df2 = df.copy()
```

```python
df2.loc[ (df2['average_rating'] >= 0) & (df2['average_rating'] <= 1), 'rating_between'] = "between 0 and 1"
df2.loc[ (df2['average_rating'] > 1) & (df2['average_rating'] <= 2), 'rating_between'] = "between 1 and 2"
df2.loc[ (df2['average_rating'] > 2) & (df2['average_rating'] <= 3), 'rating_between'] = "between 2 and 3"
df2.loc[ (df2['average_rating'] > 3) & (df2['average_rating'] <= 4), 'rating_between'] = "between 3 and 4"
df2.loc[ (df2['average_rating'] > 4) & (df2['average_rating'] <= 5), 'rating_between'] = "between 4 and 5"
```

```python
rating_df = pd.get_dummies(df2['rating_between'])
language_df = pd.get_dummies(df2['language_code'])
```

```python
features = pd.concat([rating_df,
                      language_df,
                      df2['average_rating'],
                      df2['ratings_count']], axis=1)
```

## recommendations code

```python

from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
features = min_max_scaler.fit_transform(features)
```

```python

model = neighbors.NearestNeighbors(n_neighbors=6, algorithm='ball_tree')
model.fit(features)
dist, idlist = model.kneighbors(features)
```

```python
def BookRecommender(book_name):
    book_list_name = []
    book_id = df2[df2['title'] == book_name].index
    book_id = book_id[0]
    for newid in idlist[book_id]:
        book_list_name.append(df2.loc[newid].title)
    return book_list_name

BookNames = BookRecommender('Harry Potter and the Half-Blood Prince (Harry Potter  #6)')
BookNames
```
