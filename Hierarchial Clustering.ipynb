> ### K-Means ++ Algorithm

K-Means ++ gives us a better way to initialize the clusters.

1. Choose any existing data point as a centroid.
2. Then, while selecting the second centroid, choose another existing data point which is at farthest distance from the previous one.
3. Now, the third centroid should be at farthest distance from both the previous centroids.

</br>

<img src='https://drive.google.com/uc?id=1GLUR-93dgBACFhUfRYUaQ3wLGtyAuZzU'  width='600'>

</br>

The only drawback is that it gives devastating results when we have an outlier present in the data.

</br>

NOTE : </br>
The initialization times for K-Means ++ could be slightly higher than the original K-Means algorithm.

</br>

> ### Silhouette Score

* The average distance of a data point to other data points within the same cluster should be minimum.
* The average distance of a data point to the data points of other clusters should be maximum.

</br>

<img src='https://drive.google.com/uc?id=13Ql3J33yjaOKSqt53l-FJ3XCBM8KazEm'  width='600'>

</br>

<img src='https://drive.google.com/uc?id=1etWcHEzR7QATGOeA_iBbYNTP0-Oz-oX1'  width='600'>

</br>

NOTE :

If $a(i) <<< b(i)$ </br>
then $\frac{a(i)}{b(i)} = 0$ </br>
Therefore, $s(i) = 1$

</br>

If $b(i) <<< a(i)$ </br>
then $\frac{b(i)}{a(i)} = 0$ </br>
Therefore, $s(i) = -1$

</br>

$s(i) \in (-1 to +1)$

> #### Q. How are we going to compute the similarity between groups of customers?

Let's say if we have two clusters then we can use - </br>

- Single linkage -</br>
We take the minimum possible distance between any two points of the clusters.

- Complete linkage -</br>
We take the farthest possible distance between any two points of the clusters.

</br>

<img src='https://drive.google.com/uc?id=1k4lvnmAZ3F8LkCrpa2bp8CU1r2HeNxPO'  width='600'>

</br>

- Average linkage -</br>
We take the average distance between all the points of one cluster to the other cluster.

- Centroid linkage -</br>
We find out the centroid points of the two clusters and then calculate the distance between them.

</br>

<img src='https://drive.google.com/uc?id=1z9HwxACwQ0tBXqnQGVwJftZ9vm6NL5js'  width='600'>

</br>

- **Ward method** -</br>
It sees that combining which two clusters is going to create minimum variance.

</br>

<img src='https://drive.google.com/uc?id=17S90_Z75dmj587c_MXpnwNJ-xrPVMs7l'  width='600'>

</br>

  - First figure out the centroid of the two clusters.
  - Then calculate the variance of the data points in the cluster. </br> $var = \sum{d_i^2}$ </br>
    Where $d_i$ is the distance of each point with the centroid.
  - When we combine these two clusters, we again find a centroid and calculate the variance.
  - Now, the variance will definitely increase. </br> So we choose the two clusters to be merged in such a way that the increase in variance is minimal.

</br>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore
import seaborn as sns

%matplotlib inline
```

```python
# reading the CSV file into pandas dataframe
custData = pd.read_csv("Cust_Spend_Data.csv")
custData.head(10)
```

```python
custDataAttr=custData.iloc[:,2:]
custDataAttr.head()
```

```python
custDataScaled=custDataAttr.apply(zscore)
custDataScaled.head(10)
```

```python
# importing seaborn for statistical plots
sns.pairplot(custDataScaled, height=2,aspect=2 , diag_kind='kde')
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=3, affinity='euclidean',  linkage='average')
```

```python
model.fit(custDataScaled)
```

```python
custDataAttr['labels'] = model.labels_
custDataAttr.head(10)

#custDataAttr.groupby(["labels"]).count()
```

```python
custDataClust = custDataAttr.groupby(['labels'])
```

```python
custDataClust.mean()
```

```python
from scipy.cluster.hierarchy import cophenet, dendrogram, linkage
```

```python
from scipy.spatial.distance import pdist  #Pairwise distribution between data points
```

```python
# cophenet index is a measure of the correlation between the distance of points in feature space and distance on dendrogram
# closer it is to 1, the better is the clustering

Z = linkage(custDataScaled, metric='euclidean', method='average')
c, coph_dists = cophenet(Z , pdist(custDataScaled))

c
```

```python
plt.figure(figsize=(10, 5))
plt.title('Agglomerative Hierarchical Clustering Dendogram')
plt.xlabel('sample index')
plt.ylabel('Distance')
dendrogram(Z, leaf_rotation=90.,color_threshold = 40, leaf_font_size=8. )
plt.tight_layout()
```

```python
# cophenet index is a measure of the correlation between the distance of points in feature space and distance on dendrogram
# closer it is to 1, the better is the clustering

Z = linkage(custDataScaled, metric='euclidean', method='complete')
c, coph_dists = cophenet(Z , pdist(custDataScaled))

c
```

```python
plt.figure(figsize=(10, 5))
plt.title('Agglomerative Hierarchical Clustering Dendogram')
plt.xlabel('sample index')
plt.ylabel('Distance')
dendrogram(Z, leaf_rotation=90.,color_threshold=90,  leaf_font_size=10. )
plt.tight_layout()
```

```python
# cophenet index is a measure of the correlation between the distance of points in feature space and distance on dendrogram
# closer it is to 1, the better is the clustering

Z = linkage(custDataScaled, metric='euclidean', method='ward')
c, coph_dists = cophenet(Z , pdist(custDataScaled))

c
```

```python
plt.figure(figsize=(10, 5))
plt.title('Agglomerative Hierarchical Clustering Dendogram')
plt.xlabel('sample index')
plt.ylabel('Distance')
dendrogram(Z, leaf_rotation=90.,color_threshold=600,  leaf_font_size=10. )
plt.tight_layout()
```
